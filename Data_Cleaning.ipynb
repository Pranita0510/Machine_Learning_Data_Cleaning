```python
# Import pandas

import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import dataset

df = pd.read_csv("data.csv")
print(df.to_string())
```

```python
df.info()
```

```python
# handling empty cells

# Using droping method
new_df = df.dropna()
print(new_df)
```

```python
new_df.info()
```

```python
# Handling empty cells

# Using fill method
df.fillna(130, inplace = True)
```

```python
df.info()
```

```python
# Handling empty cells of specific column

df["Calories"].fillna(130, inplace = True)
```

```python
df.info()
```

```python
# Handling empty cells

# Using mean

x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True)
```

```python
df.info()
```

```python
# Using median

x = df["Calories"].median()
df["Calories"].fillna(x, inplace = True)
```

```python
df.info()
```

```python
# Using mode

x = df["Calories"].mode()[0] 
df["Calories"].fillna(x, inplace = True) 
```

```python
df.info()
```
